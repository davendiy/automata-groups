## Randomness investigation

```python
def random_el(n):
    space = ['a', 'b', 'c']
    first = np.random.choice(space)
    res = [first]
    for i in range(1, n):
        if res[i-1] == 'a':
            
            res.append(np.random.choice(['b', 'c']))
        elif res[i-1] == 'b':
            res.append(np.random.choice(['a', 'c']))
        else:
            res.append(np.random.choice(['a', 'b']))
    return ''.join(res)
```

### Check average size growth

```python
av_sizes = []

for n in range(2, 7):
    av = 0
    for _ in range(50):
        av += from_string(random_el(2**n)).tree.size()
    av_sizes.append(av / i)
    print(2**n, av/i)
```

## Distribution of $w_1$

```python
def permute(seq, repeat):
    if repeat == 1:
        for el in seq:
            yield [el]
    elif repeat < 1:
        yield []
    else:
        for prev in permute(seq, repeat-1):
            for el in seq:
                if prev[-1] == el:
                    continue
                yield prev + [el]

x = []
x_lengths = []

for i in range(1, 13):
    for el in permute(['a', 'b', 'c'], repeat=i):
        tmp = from_string(el)
        x.append(tmp[0].name)
        x_lengths.append(len(tmp[0].name))
```

```python
from collections import Counter

dictionary = {}
y = []

counter = Counter(x)
for i, el in enumerate(counter):
    dictionary[el] = i
    y += [i] * counter[el]
plt.hist(y, bins=80)
```
