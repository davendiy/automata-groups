### Generating Hanoi Tower Group on 3 pegs without reduce function (i.e. words like 'aaabbccac' won't be reduced into 'c')

```python
H3 = AutomataGroup.generate_H3()
H3
```

### Let's get some random element of length 20

```python
x = H3.random_el(20)
x
```

### We can get the first level sections

```python
x[0], x[1], x[2]
```

### as well as the second level sections

```python
x[0][1]
```

### as well as slice of any length 

```python
x[0, 1, 1]
```

### Also we can draw the entire tree-like structure

```python
x.show()
```

### and get the full description of this element, applying algorithm that solves the order problem 

```python
x.describe(DiGraph)
```

### Let's generate some other element and describe it. 

```python
y = H3.random_el(20)
y.describe(DiGraph)
```

### Multiplication within the group is straightforward

```python
x * y
```

```python
(x * y).show()
```

### as well as checking elements' properties.

```python
x.is_one()
```

```python
x.order()
```

```python
particular_el = H3('abcabcabc')
particular_el.show()
```

### Multiplication is very fast by the way. 

```python
x ** 20
```
