```python
import numpy as np
from src.source2 import *
%matplotlib inline

def random_el(n):
    space = ['a', 'b', 'c']
    first = np.random.choice(space)
    res = [first]
    for i in range(1, n):
        if res[i-1] == 'a':
            
            res.append(np.random.choice(['b', 'c']))
        elif res[i-1] == 'b':
            res.append(np.random.choice(['a', 'c']))
        else:
            res.append(np.random.choice(['a', 'b']))
    return ''.join(res)
```

```python
tests = 1000
n = 30

def random_sons(n, tests=10_000):

    first_layer = []
    second_layer = []

    for i in range(tests):
        print(f"\rDone: {str(i/tests * 100)[:5].rjust(6)}%".ljust(50), end='')
        tmp_el = from_string(random_el(n))
        try:
            first_layer.append(tmp_el[0].name)
            second_layer.append(tmp_el[0][0].name)
        except IndexError:
            print("\rBad element:", tmp_el)
    return first_layer, second_layer

first_layer, second_layer = random_sons(n, tests)
```

```python
from collections import Counter
counter_first = Counter(first_layer)
counter_second = Counter(second_layer)
counter_first.most_common(10)
```

```python
def hist_counter(counter):
    res = []
    dictionary = {}
    for i, (name, amount) in enumerate(counter.most_common()):
        res += [i] * amount
        dictionary[i] = name
    return res, dictionary

hist_first, dict_first = hist_counter(counter_first)
plt.hist(hist_first, bins=100)
plt.title('Distribution of W_1 (from the most common to the least)')
plt.show()

hist_second, dict_second = hist_counter(counter_second)
plt.hist(hist_second, bins=100)
plt.title('Distribution of W_2 (from the most common to the least)')
plt.show()
```

```python
def investigate_sons(n, tests):
    print("Start random generating...")
    first_layer, second_layer = random_sons(n, tests)
    
    counter_first = Counter(first_layer)
    counter_second = Counter(second_layer)
    fig, axs = plt.subplots(2, 2, figsize=(12, 12))
    hist_first, dict_first = hist_counter(counter_first)
    axs[0, 0].hist(hist_first, bins=100)
    axs[0, 0].title.set_text('W_1 (from the most common to the least)')

    hist_second, dict_second = hist_counter(counter_second)
    
    axs[0, 1].hist(hist_second, bins=100)
    axs[0, 1].title.set_text('W_2 (from the most common to the least)')
    
    axs[1, 0].hist(list(map(len, first_layer)), bins=100)
    axs[1, 0].title.set_text('Lenghts of W_1')
    
    axs[1, 1].hist(list(map(len, second_layer)), bins=100)
    axs[1, 1].title.set_text('Lengths of W_1_1')
    plt.show()
```

```python
investigate_sons(30, 1000)
```

```python
from itertools import product


n = 8
sizes_w1 = []
sizes_w2 = []
sizes_w3 = []
sizes_ws = []
sizes_w11 = []

i = 0
amount = 3 ** n

for el in product('abc', repeat=n):
    tmp = from_string(el)
    try:
        x1 = len(tmp[0].name) if tmp[0].name != 'e' else 0
        x2 = len(tmp[1].name) if tmp[1].name != 'e' else 0
        x3 = len(tmp[2].name) if tmp[2].name != 'e' else 0
        sizes_w1.append(x1)
        sizes_w2.append(x2)
        sizes_w3.append(x3)
        sizes_ws.append((x1, x2, x3))
        
#         x11 = 
#         sizes_w11.append(len(tmp[0][0].name))
    except Exception:
        continue
    print('\r', end='')
    print(f'{str((i / amount) * 100)[:5].rjust(6)}%', end='')
    i += 1
```
