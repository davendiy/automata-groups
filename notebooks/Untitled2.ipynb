```python
from itertools import product
def entire_sum(n):
    res = 0
    i = 0
    all_amount = 3 ** n
    for el in product("abc", repeat=n):
        print(f'\r{str(i / all_amount * 100)[:4].rjust(6)}%', end='')
        res += from_string(el).tree.size()
        i += 1
    return res

entire_sum(6)
```

```python
for n in range(6, 15):
    print(f"\nn = {n}...")
    res = entire_sum(n)
    print(f"\nres: {res}")
```

```python
from scipy.special import binom
import numpy as np
import matplotlib.pyplot as plt


def dist(k, n):
    res = (2 ** (n - k - 1) * binom(n-1, k)) / 3**(n-1)
    return res


T(n, k) = 2 T(n-1, k) + T(n-1, k-1)

n = 100

x = np.arange(0, n, dtype='float64')
y = dist(x, n)
plt.plot(x, y)
print(y.sum())
```

```python
def permute(seq, repeat):
    if repeat == 1:
        for el in seq:
            yield [el]
    elif repeat < 1:
        yield []
    else:
        for prev in permute(seq, repeat-1):
            for el in seq:
                if prev[-1] == el:
                    continue
                yield prev + [el]


min_els = []
for n in range(10):
    min_size = float('inf')
    min_el = ''
    for el in permute('abc', repeat=n):
        tmp_size = from_string(el).tree.size()
        if tmp_size < min_size:
            min_el = ''.join(el)
    min_els.append(min_el)
```
