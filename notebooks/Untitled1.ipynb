```python
import numpy as np
from functools import lru_cache


def f(n):
    return n * np.floor(np.log2(n) + 1) + 2*n - 2 ** (np.floor(np.log2(n)) + 1)


@lru_cache(maxsize=32)
def f_recursively(n):
    if n == 1:
        return 1
    if n == 0:
        return 0
    return f_recursively(np.floor(n / 2)) + f_recursively(np.ceil(n / 2)) + n

n = 5
print(f(n), f_recursively(n))
```

```python
def test(n):
    return f_recursively(n+1) - f_recursively(n)

def test2(n):
    return np.floor(np.log2(n)) + 3

n = 11100000
print(test(n), test2(n))
```
