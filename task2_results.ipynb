```python
def permute(seq, repeat):
    if repeat == 1:
        for el in seq:
            yield [el]
    elif repeat < 1:
        yield []
    else:
        for prev in permute(seq, repeat-1):
            for el in seq:
                if prev[-1] == el:
                    continue
                yield prev + [el]

max_size = np.argmax(tmp_sizes)
max_amounts = np.argmax(tmp_amounts)
                

for j, el in enumerate(permute('abc', repeat=20)):
    if j == max_size:
        max_size_el = el
    if j == max_amounts:
        max_amounts_el = el
```

```python
max_size_tree = from_string(max_size_el)
print(f"size: {max_size_tree.tree.size()}",
      f"height: {max_size_tree.tree.height()}",
      f"vertex amount: {max_size_tree.tree.vert_amount()}",
      sep='\n')
max_size_tree.show(y_scale_mul=4, lbn=True)
```

```python
max_amounts_tree = from_string(max_amounts_el)
print(f"size: {max_amounts_tree.tree.size()}",
      f"height: {max_amounts_tree.tree.height()}",
      f"vertex amount: {max_amounts_tree.tree.vert_amount()}",
      sep='\n')
max_amounts_tree.show(y_scale_mul=4, lbn=True)
```
